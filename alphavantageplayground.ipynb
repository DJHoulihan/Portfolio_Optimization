```python
import requests
import pandas as pd

HEADERS = {
    "Accept-Encoding": "gzip, deflate"
}

def fetch_indicator(symbol, function, indicator_name, **params):
    params.update({
        "function": function,
        "symbol": symbol,
        "apikey": ALPHA_VANTAGE_API_KEY
    })

    r = requests.get(
        AV_API_URL,
        params=params,
        headers=HEADERS,
        timeout=30
    )

    r.raise_for_status()
    data = r.json()

    tech_key = next(k for k in data if k.startswith("Technical Analysis"))
    df = pd.DataFrame.from_dict(data[tech_key], orient="index")

    df.index = pd.to_datetime(df.index)
    df = df.rename(columns={df.columns[0]: "value"})
    df["indicator"] = indicator_name

    return df.reset_index().rename(columns={"index": "date"})
```

```python
symbol = "AAPL"

rsi = fetch_indicator(
    symbol,
    function="RSI",
    indicator_name="RSI",
    interval="daily",
    time_period=14,
    series_type="close"
)

sma = fetch_indicator(
    symbol,
    function="SMA",
    indicator_name="SMA_20",
    interval="daily",
    time_period=20,
    series_type="close"
)

ema = fetch_indicator(
    symbol,
    function="EMA",
    indicator_name="EMA_20",
    interval="daily",
    time_period=20,
    series_type="close"
)
```
